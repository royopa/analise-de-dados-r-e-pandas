### Aula 2 - Criação de gráficos

#### para instalar as bibliotecas execute:
##### install.packages('readr')
##### install.packages("plyr")

```r
library(readr)
library(plyr)
```

```r
duracao <- read_csv('datasets/duracao.csv')
```

```r
head(duracao)
```

```r
duracao <- rename(duracao, replace = c('user_id'='aluno', 'course_id'='curso', 'timeToFinish'='dias'))
```

```r
head(duracao)
```

```r
plot(duracao$dias)
```

```r
hist(duracao$dias)
```

```r
hist(duracao$dias, breaks=3)
```

```r
hist(duracao$dias, breaks=20)
```

```r
hist(duracao$dias, breaks = 20, main = "Histograma do Tempo")
```

```r
hist(duracao$dias, breaks = 20, main = "Histograma do Tempo", ylab = "Quantidade", xlab = "Tempo", ylim = c(0,2500), col = "blue")
```
